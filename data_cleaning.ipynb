```python
import pandas as pd 
import numpy as np
```

```python
df_example = pd.read_csv('audible_uncleaned.csv')
df_example
```

```python
## AUTHOR
df_example["author"].str.lstrip("Writtenby:")
df_example["author"] = df_example["author"].str.lstrip("Writtenby:")
df_example

```

```python
import matplotlib.pyplot as plt #importing visualization package

df_example["author"].str.lstrip("Writtenby:")
df_example["author"] = df_example["author"].str.lstrip("Writtenby:")

valueCounts = df_example['author'].head(100).value_counts()

valueCounts.plot(kind='bar')
```

```python
## NARRATOR
df_example["narrator"] = df_example["narrator"].str.lstrip("Narratedby:")
valueCounts = df_example['narrator'].head(100).value_counts()

valueCounts.plot(kind='bar')
```

```python
## TIME
#2 hrs and 10 minutes
def convertTime(timeString):
    totalMins = 0
    splitString = timeString.split(' and ')
    for item in splitString:
        if 'hrs' in item:
            hours = int(item.split(' ')[0])
            totalMins += (60*hours)
        elif 'mins' in item:
            mins = int(item.split(' ')[0])
            totalMins += mins
    return totalMins

df_example['time'] = df_example['time'].apply(convertTime) ##how to apply a function to a colum pandas https://stackoverflow.com/questions/34962104/how-can-i-use-the-apply-function-for-a-single-column
df_example
```

```python
## RELEASE DATE

df_example['releasedate'] = pd.to_datetime(df_example["releasedate"], infer_datetime_format=True, format='mixed').dt.strftime('%Y-%m-%d')
df_example
```

```python
## STARS

notRated = df_example[df_example['stars'] == 'Not rated yet'] #no rating
rated = df_example[df_example['stars'] != 'Not rated yet'] #rating

rated
```

```python
rated[['stars', 'numRatings']] = rated['stars'].str.split('stars', expand = True)
rated
```

```python
rated['stars'] = rated['stars'].str.split(" out of ").str[0]
rated['stars'].str.split(" out of ").str[0]

df_cloned = pd.concat([rated,notRated])

df_cloned 
```
